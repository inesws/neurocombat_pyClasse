# Explaning ComBatHarmonization

```python
"""
------ComBatHarmonization Attributes--------------------------------------------

 cv_method=None : this options is not finished

 ref_batch= ref_batch: if ref_batch exist, is an integer corresponding to batch id, otherwise ref_batch=None

 regression_fit: whether the data will be used in the BiocovRegression afterwards.
                 0 -> outputs the harmonized data only array (n_subj x n_feat);
                 1 -> outputs a dictionary of the form of the input with harmonize
                 data and respective covariates {'data': harmonize_output ; 'covariates': covars_matrix}
                 ready for the linear regression fit.

 feat_detail: The details of the features to harmonize which are organized in a dictionary of dictionaries
              Each first level of the dictionary corresponds to a set of features of the same name, e.g: Cortical thickness
              , Volumes, surface areas, curvature; for each of this first level a second within dictionary is defined with
              the column id of those features in the dataset 'id' and the corresponding biocovariates to preserve
              in case there is none, define 'categorical' or 'continuous' with a empty list (eg: 'categorical': [] )

              dict= {'name_of_feature_type_1: {'id': columns_id_corresponding,
              'categorical': list_of_names_of_column_in_dataframe_of_categorical_covaariates,
              'continuous': list_of_names_of_column_in_dataframe_of_continuous_covaariates }

              -each of the levels of feature_type_i is harmonized independently of the others;

feat_of_no_interest: better not to use this option for now, but it was to give
                     the covariates to harmonize before harmonizing the main features.
-------Input--------------------------------------------------------------------

dictionary_input: a dictionary with 2 keys, 'data' and 'covariates'.
                  'data': a dataframe (n_subjxn_feat)
                  'covariates' a dataframe (n_subj x n_covariates) -> the columns
                  in here need to have the names of biocovariates consistent with
                  the ones in feat_detail 'continuous' and 'categorical'. It needs
                  also the batch columns with the batch id's as integers! The batch
                  columns needs to be named exactly 'batch' (see above strat_covars
                  dataframe for a concrete example).
------Ouput---------------------------------------------------------------------
Depending on the regression_fit option:

regression_fit: 0 -> outputs the harmonized data only array (n_subj x n_feat);
                 1 -> outputs a dictionary of the form of the input with harmonize
                 data and respective covariates {'data': harmonize_output ; 'covariates': covars_matrix}

"""
# feat_detail is the information related to the features to harmonized:
# corresponding columns in dataset and respective biocovariates to consider

cortical_columns= np.arange(0,68) # 68 cortical thickness DK40
gvolumes_columns=np.arange(68,118) # 50 gray matter volumes CoBraLab
wvolumes_columns=np.arange(118,170) # 52 white mater volumes CoBraLab
volumes_columns=np.concatenate((gvolumes_columns,wvolumes_columns)) # all volumes

feat_detail={'cortical_thickness': {'id': cortical_columns,
                                                   'categorical': ['Gender'],
                                                   'continuous':['Age']},
              'volumes':           {'id': volumes_columns,
                                        'categorical': ['Gender'],
                                        'continuous':['Age', 'harmonized_TIV']}}

define_combat_function=ComBatHarmonization(cv_method=None, ref_batch=None,
                                           regression_fit=0,
                                           feat_detail=feat_detail,
                                           feat_of_no_interest=None)
```

# Explaining BiocovRegression

```python
"""
------BiocovRegression Attributes--------------------------------------------

 cv_method=None : this options is not finished

 ref_batch= ref_batch: if ref_batch exist, is an integer corresponding to batch id, otherwise ref_batch=None
                       In case reference is defined, the LR is only fitted in the reference data.

 output_dictionary: False -> outputs the corrected data only array (n_subj x n_feat);
                    True -> outputs a dictionary of the form of the input with corrected
                    data and respective covariates {'data': corrected_output ; 'covariates': covars_matrix}

 feat_detail: The details of the features to correct which are organized in a dictionary of dictionaries
              similarly to the ComBatHarmonization.
              Each first level of the dictionary corresponds to a set of features of the same name, e.g: Cortical thickness
              , Volumes, surface areas, curvature; for each of this first level a second within dictionary is defined with
              the column id of those features in the dataset 'id' and the corresponding biocovariates to remove
              in case there is none, define 'categorical' or 'continuous' with a empty list (eg: 'categorical': [] )

              dict= {'name_of_feature_type_1: {'id': columns_id_corresponding,
              'categorical': list_of_names_of_column_in_dataframe_of_categorical_covaariates,
              'continuous': list_of_names_of_column_in_dataframe_of_continuous_covaariates }

              -each of the levels of feature_type_i uses different biocovariates in the LR;

feat_of_no_interest: better not to use this option for now, but it was to give
                     the covariates to harmonize before harmonizing the main features.
-------Input--------------------------------------------------------------------

dictionary_input: a dictionary with 2 keys, 'data' and 'covariates'.
                  'data': a dataframe (n_subjxn_feat)
                  'covariates' a dataframe (n_subj x n_covariates) -> the columns
                  in here need to have the names of biocovariates consistent with
                  the ones in feat_detail 'continuous' and 'categorical'. It needs
                  also the batch columns with the batch id's as integers! The batch
                  columns needs to be named exactly 'batch' (see above strat_covars
                  dataframe for a concrete example).
------Ouput---------------------------------------------------------------------
Depending on the output_dictionary option:

output_dictionary: False -> outputs the corrected data only array (n_subj x n_feat);
                    True -> outputs a dictionary of the form of the input with corrected
                    data and respective covariates {'data': corrected_output ; 'covariates': covars_matrix}

"""


biocovariates_linear_regression=BiocovariatesRegression(ref_batch=None,
                                                        cv_method=None,
                                                        feat_detail=feat_detail,
                                                        output_dict=True)
```

# Explaining StandardScalerDict

```python
"""
StandardScalerDict is a standardscaler class created to be compatible with
ComBatHarmonization and BiocovRegression where we can define a ref_batch to which
to fit the scaler and can choose whether to standardiza 'data' and 'covariates'
in the dictionary or just 'data' or just 'covariates'

------StandardScalerDict Attributes--------------------------------------------


 ref_batch= ref_batch: if ref_batch exist, is an integer corresponding to batch id, otherwise ref_batch=None
                       In case reference is defined, the scaler is only fitted in the reference data.

 output_dictionary: False -> outputs the corrected data only array (n_subj x n_feat);
                    True -> outputs a dictionary of the form of the input with corrected
                    data and respective covariates {'data': corrected_output ; 'covariates': covars_matrix}

 std_data: True/False -> standardize feature data (n_subj x n_feat)
 std_cov : True/False -> standardize biocovariates (n_subj x n_cov)

output_dictionary: False -> outputs the standardize data only array (n_subj x n_feat);
                    True -> outputs a dictionary of the form of the input with corrected
                    data and respective covariates {'data': output_data ; 'covariates': output_covars}
-------Input--------------------------------------------------------------------

dictionary_input: a dictionary with 2 keys, 'data' and 'covariates'.
                  'data': a dataframe (n_subjxn_feat)
                  'covariates' a dataframe (n_subj x n_covariates) -> the columns
                  in here need to have the names of biocovariates consistent with
                  the ones in feat_detail 'continuous' and 'categorical'. It needs
                  also the batch columns with the batch id's as integers! The batch
                  columns needs to be named exactly 'batch' (see above strat_covars
                  dataframe for a concrete example).
------Ouput---------------------------------------------------------------------
Depending on the output_dictionary option:

output_dictionary: False -> outputs the standardize data only array (n_subj x n_feat);
                    True -> outputs a dictionary of the form of the input with corrected
                    data and respective covariates {'data': output_data ; 'covariates': output_covars}

"""

standard_scaler_function=StandardScalerDict(ref_batch=None,
                                            std_data=True,
                                            std_cov=False,
                                            output_dict=True)
```

# Defining a Pipeline with sklearn with the Confounder Removal Classes

```python
"""
----- Steps of the Pipeline in orde--------------------------------------------

1. ComBat Harmonization -> output dictionary 'data' and 'covariates;
2. Standard Scaler -> standardize only 'data'; output dictionary
3. BiocovariatesRegression -> output only data ;
4. Standard Scaler -> standardized adjusted data ;

"""

combat_function=ComBatHarmonization(cv_method=None, ref_batch=None,
                                    regression_fit=1,
                                    feat_detail=feat_detail,
                                    feat_of_no_interest=None)


scaler_data_fun_1=StandardScalerDict(ref_batch=None,
                                     std_data=True,
                                     std_cov=False,
                                     output_dict=True)

biocov_regre=BiocovariatesRegression(ref_batch=None,
                                     cv_method=None,
                                     feat_detail=feat_detail,
                                     output_dict=True)

scaler_data_fun_2=StandardScalerDict(ref_batch=None,
                                     std_data=True,
                                     std_cov=False,
                                     output_dict=False)

confounder_removal_steps = [
    ('Multi-center features Harmonization', combat_function),
    ('Standardize only data 1', scaler_data_fun_1),
    ('Linear Confounder Regression', biocov_regre),
    ('Standardize only data 2', scaler_data_fun_2)]

CR_pipeline = Pipeline(confounder_removal_steps)
```

# Harmonize Total Intracranial Volume

```python

# M-ComBat: If there is a reference set : ref_batch

globals_id=np.arange(0,6)

# The TIV is harmonized together with globals
globals_of_interest={'globals': {'id': globals_id, 'categorical': ['Gender'],
                               'continuous':['Age']} }

# Define combat object
combat_function_TIV=ComBatHarmonization(cv_method=None, ref_batch=None,
                                        regression_fit=0,
                                        feat_detail=globals_of_interest,
                                        feat_of_no_interest=None)

# Harmonize TIV

# Define input
train_globals_dict={'data': globals_train, 'covariates': covars_train}

# Harmonize Training Globals Data: fit and transform
globals_train_harm=combat_function_TIV.fit_transform(train_globals_dict)

# Harmonize Testing Globals Data: fit and transform

test_globals_dict={'data': globals_test, 'covariates': covars_test}

globals_test_harm=combat_function_TIV.transform(test_globals_dict)

# Extrat harmonized TIV
train_harm_TIV=pd.DataFrame(globals_train_harm[:,0], columns=['TIV'])
test_harm_TIV=pd.DataFrame(globals_test_harm[:,0], columns=['TIV'])

# Replace raw TIV covariate with harmonized TIV
covars_train['TIV']=globals_train_harm[:,0]
covars_test['TIV']=globals_test_harm[:,0]
```

# Harmonize Brain Morphological Data

```python

# Define inputs
train_strat_dict={'data': X_train, 'covariates': covars_train}
test_strat_dict={'data': X_test, 'covariates': covars_test}

# Brain feature-types to consider separatly and associated bio-covariates
features_info={'cortical': {'id': cortical_columns,
                               'categorical': ['Gender'],
                               'continuous':['Age']},
                  'volumes': {'id': volumes_columns,
                              'categorical': ['Gender'],
                              'continuous':['Age','TIV']}}

# Confounder Removal Pipelne

combat_data=ComBatHarmonization(cv_method=None, ref_batch=None,
                                    regression_fit=1,
                                    feat_detail=features_info,
                                    feat_of_no_interest=None)

scaler_data_fun_1=StandardScalerDict(ref_batch=None,std_data=True,
                                     std_cov=False,
                                     output_dict=True)

biocov_regre=BiocovariatesRegression(ref_batch=None,cv_method=None,
                                     feat_detail=features_info,
                                     output_dict=True)

scaler_data_fun_2=StandardScalerDict(ref_batch=None,std_data=True,
                                     std_cov=False,
                                     output_dict=False)

combat_biocovregr_steps = [('Multi-center features Harmonization', combat_data),
 ('Standardize only data 1', scaler_data_fun_1),('Confounder Regression', biocov_regre),
  ('Standardize only data 2', scaler_data_fun_2)]

preprocessing_pipeline = Pipeline(combat_biocovregr_steps)

# Fit_transform method in training set
X_train_corr=preprocessing_pipeline.fit_transform(train_strat_dict)

# Transforming Test set
X_test_corr=preprocessing_pipeline.transform(test_strat_dict)
```
